```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Carregar um arquivo CSV contendo dados de vendas em um DataFrame. Realizar a limpeza dos dados, removendo valores nulos e formatando os tipos de dados corretamente.
file = "./Database/vendasMercado.csv"
dados_df = pd.read_csv(file)
dados_df
```

```python
dados_df.info()
```

```python
dados_df["Data"] = pd.to_datetime(dados_df["Data"], format = ("%m/%d/%Y"))
dados_df["Hora"] = pd.to_datetime(dados_df["Hora"], format = ("%H:%M")).dt.time
```

```python
dados_df
```

```python
# 2. Calcular a média, mediana e desvio padrão das vendas totais.
display(f"Média: ........... {dados_df['Total'].mean()}")
display(f"Médiana: ......... {dados_df['Total'].median()}")
display(f"Desvio Padrão : .. {dados_df['Total'].std()}")
```

```python
dados_df.describe().round(2)
```

```python
# 3. Identificar as 3 linhas  de produtos mais vendidos e plotar um gráfico de barras mostrando a quantidade de vendas de cada um.
dados_df["Linha de Produto"].unique()
```

```python
dados_df["Quantidade"].sum()
```

```python
linha_df = dados_df.groupby("Linha de Produto")["Quantidade"].sum()
linha_df
```

```python
linha_df = pd.DataFrame(linha_df.sort_values(ascending=False).head(3).reset_index())
linha_df
```

```python
sns.barplot(linha_df, x ="Linha de Produto", y = "Quantidade")
plt.title("Top quantidade por categoria")
plt.show()
```

```python
# 4. Calcular o total de vendas por cidade e plotar um gráfico de pizza mostrando a distribuição das vendas.
dados_df["Cidade"].unique()
```

```python
cidade_vendas_df = pd.DataFrame(dados_df.groupby("Cidade")["Quantidade"].sum().reset_index())
cidade_vendas_df
```

```python
plt.figure(figsize=(8, 6))
plt.title('Distribuição das Vendas por Região')
plt.pie(cidade_vendas_df["Quantidade"].tolist(), labels=cidade_vendas_df["Cidade"].to_list(),autopct='%1.1f%%')


plt.show()
```

```python
# 5. Criar uma nova coluna no DataFrame que represente o media gasta por produto com imposto.
dados_df["Media Produto Total"] = dados_df["Total"] / dados_df["Quantidade"] 
dados_df.head(5)
```

```python
# 6. Calcular o imposto total por mês e plotar um gráfico de linha mostrando a evolução do imposto ao longo do tempo.
mes_df = dados_df.groupby(pd.Grouper(key='Data', freq='m'))["Impostos"].sum()
mes_df
```

```python
sns.set(style="darkgrid")
sns.lineplot(mes_df)
plt.xticks(mes_df.index)
plt.yticks(np.arange(start=1000, stop = 8001, step = 1000))
plt.yscale("linear")
plt.show()
```

```python
# 7. criar dois novos DataFrames com clientes separados pelo sexo.
display(dados_df["Genero"].unique())
df_mulheres = dados_df.query("Genero == 'Female'")
df_mulheres.head(5)
```

```python
# 8. Calcular a correlação entre o preço de venda e a quantidade vendida de cada produto.
correlacao_df = dados_df["Total"].corr(dados_df["Quantidade"])
correlacao_df
```

```python
sns.relplot(dados_df, x = "Total", y = "Quantidade")
```

```python
# 9. Agrupar as vendas por categoria de produto e calcular a média e o desvio padrão das vendas para cada categoria.
dados_df.groupby("Linha de Produto")["Total"].agg(["mean","std"]).reset_index()
```

```python
# 10. Calcular o valor total das vendas do ultimo mês, e comparar com o mês anterior.
fevereiro_df = dados_df.query("Data >= '2019-02-01'  & Data < '2019-03-01'")["Total"].sum()
marco_df = dados_df.query("Data >= '2019-01-01'  & Data < '2019-02-01'")["Total"].sum()
display(f"Fevereiro: ....: {fevereiro_df}")
display(f"Março: ........: {marco_df}")
```

```python
meses = ["Fev", "Mar"]

plt.bar(meses,(fevereiro_df, marco_df))
plt.title("Venda Total Fev/Mar")
plt.show()
```
