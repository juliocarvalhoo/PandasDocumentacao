```python
import pandas as pd
import numpy as np
```

```python
# 1. Carregue um arquivo CSV chamado "dados.csv" em um DataFrame.
file = "./Database/customer_shopping_data .csv"

dados_df = pd.read_csv(file)
```

```python
# 2. Verifique as primeiras 5 linhas do DataFrame.
dados_df.head()
```

```python
# 3. Exiba as informações estatísticas básicas (contagem, média, desvio padrão, etc.) das colunas numéricas do DataFrame.
dados_df.describe().round(2)
```

```python
# 4. Selecione apenas as linhas do DataFrame onde a coluna "idade" seja maior que 30.
idadeMaior30_df = dados_df.query("age > 30")
idadeMaior30_df
```

```python
jeito2_df = dados_df.where(dados_df["age"] > 30).dropna()
jeito2_df
```

```python
# 5. Calcule a média da coluna "price".
dados_df["price"].mean()

```

```python
# 6. Ordene o DataFrame pela coluna "data" em ordem decrescente 16/05/2021.
#covnertendo para o formato datetime
dados_df["invoice_date"] = pd.to_datetime(dados_df["invoice_date"], format="%d/%m/%Y")

dados_df.sort_values("invoice_date", ascending= False)
```

```python
# 7. Crie uma nova coluna chamada "dobro_preco" que contenha o dobro do valor da coluna "price".

dados2_df = dados_df
dados2_df["dobro_preco"] = dados2_df["price"] * 2
dados2_df
```

```python
# 8. Remova a coluna "gênero" do DataFrame.
del dados_df["gender"]
```

```python
# 9. Agrupe o DataFrame por "payment_method" e calcule a soma dos gastos por método.
dados_payment_df = dados_df.groupby("payment_method")["price"].sum()
dados_payment_df
```

```python
# 10. Exporte o DataFrame resultante para um novo arquivo CSV chamado "dados_processados.csv".
dados_df.to_csv("./Database/dados_processados.csv", index = True)
```
